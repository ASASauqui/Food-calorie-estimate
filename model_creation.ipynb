## ==================== Libraries ====================

```python
# General.
import numpy as np

# Data processing.
import os

# Data visualization.
import matplotlib.pyplot as plt

# Models.
from torchvision.datasets import ImageFolder
```

## ==================== Load Data ====================

### Paths

```python
training_data_dir = './Datasets/Training/'
test_data_dir = './Datasets/Test/'
```

### Load data

```python
training_data = ImageFolder(root=training_data_dir)
test_data = ImageFolder(root=test_data_dir)
print(training_data, test_data)
```

### Show images

```python
plt.figure(figsize=(8,4))
training_index = np.random.randint(0, len(training_data))
test_index = np.random.randint(0, len(test_data))
plt.subplot(1,2,1)
plt.title('Random image (TRAINING DATA)', fontsize=12)
plt.imshow(training_data[training_index][0])
plt.subplot(1,2,2)
plt.title('Random image (TEST DATA)', fontsize=12)
plt.imshow(test_data[test_index][0])
```

### Count training data

```python
# Classes array.
class_names = []

# Dictionary to count the data.
training_class_count = {}

# Iterate through each image within the training dataset.
for img in training_data.imgs:
    # Get class name.
    class_name = img[0].split('/')[3].split('\\')[0]
    
    # If that name is not in the array.
    if class_name not in class_names:
        class_names.append(class_name)

    # Get class index.
    index_class = img[1]
    
    # If that key is not in the dictionary.
    if index_class not in training_class_count.keys():
        training_class_count[index_class] = 0

    # Increment class counter.
    training_class_count[index_class] += 1

# Print the count.
for i in training_class_count:
    print(class_names[i] + ': ' + str(training_class_count[i]))
```

### Count test data

```python
# Dictionary to count the data.
test_class_count = {}

# Iterate through each image within the test dataset.
for img in test_data.imgs:
    # Get class index.
    index_class = img[1]
    
    # If that key is not in the dictionary.
    if index_class not in test_class_count.keys():
        test_class_count[index_class] = 0

    # Increment class counter.
    test_class_count[index_class] += 1

# Print the count.
for i in test_class_count:
    print(class_names[i] + ': ' + str(test_class_count[i]))
```

### Show distribution

```python
plt.figure(figsize=(15,10))

plt.subplot(2,1,1)
plt.title('Count of images (TRAINING DATA)', fontsize=14)
for i, _ in enumerate(training_class_count):
    plt.bar(class_names[list(training_class_count.keys())[i]],
            list(training_class_count.values())[i]);

plt.subplot(2,1,2)
plt.title('Count of images (TEST DATA)', fontsize=14)
for i, _ in enumerate(test_class_count):
    plt.bar(class_names[list(test_class_count.keys())[i]],
            list(test_class_count.values())[i]);

plt.show()
```
