## ==================== Libraries ====================

```python
# General.
import numpy as np

# Data processing.
import os

# Data visualization.
import matplotlib.pyplot as plt

# Models.
from torchvision.datasets import ImageFolder
```

## <br><br><br><br><br>==================== Load Detection Data ====================

### Paths

```python
detection_training_data_dir = './Datasets/Detection/Training/'
detection_test_data_dir = './Datasets/Detection/Test/'
```

### Load data

```python
detection_training_data = ImageFolder(root=detection_training_data_dir)
detection_test_data = ImageFolder(root=detection_test_data_dir)
print(detection_training_data, detection_test_data)
```

### Show images

```python
plt.figure(figsize=(8,4))
detection_training_index = np.random.randint(0, len(detection_training_data))
detection_test_index = np.random.randint(0, len(detection_test_data))
plt.subplot(1,2,1)
plt.title('Random image (TRAINING DATA)', fontsize=12)
plt.imshow(detection_training_data[detection_training_index][0])
plt.subplot(1,2,2)
plt.title('Random image (TEST DATA)', fontsize=12)
plt.imshow(detection_test_data[detection_test_index][0])
```

### Count training data

```python
# Classes array.
class_names = []

# Dictionary to count the data.
detection_training_class_count = {}

# Iterate through each image within the training dataset.
for img in detection_training_data.imgs:
    # Get class name.
    class_name = img[0].split('/')[4].split('\\')[0]
    
    # If that name is not in the array.
    if class_name not in class_names:
        class_names.append(class_name)

    # Get class index.
    index_class = img[1]
    
    # If that key is not in the dictionary.
    if index_class not in detection_training_class_count.keys():
        detection_training_class_count[index_class] = 0

    # Increment class counter.
    detection_training_class_count[index_class] += 1

# Print the count.
for i in detection_training_class_count:
    print(class_names[i] + ': ' + str(detection_training_class_count[i]))
```

### Count test data

```python
# Dictionary to count the data.
detection_test_class_count = {}

# Iterate through each image within the test dataset.
for img in detection_test_data.imgs:
    # Get class index.
    index_class = img[1]
    
    # If that key is not in the dictionary.
    if index_class not in detection_test_class_count.keys():
        detection_test_class_count[index_class] = 0

    # Increment class counter.
    detection_test_class_count[index_class] += 1

# Print the count.
for i in detection_test_class_count:
    print(class_names[i] + ': ' + str(detection_test_class_count[i]))
```

### Show distribution

```python
plt.figure(figsize=(15,10))

plt.subplot(2,1,1)
plt.title('Count of images (TRAINING DATA)', fontsize=14)
for i, _ in enumerate(detection_training_class_count):
    plt.bar(class_names[list(detection_training_class_count.keys())[i]],
            list(detection_training_class_count.values())[i]);

plt.subplot(2,1,2)
plt.title('Count of images (TEST DATA)', fontsize=14)
for i, _ in enumerate(detection_test_class_count):
    plt.bar(class_names[list(detection_test_class_count.keys())[i]],
            list(detection_test_class_count.values())[i]);

plt.show()
```
